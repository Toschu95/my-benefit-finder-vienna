```python
import requests
import json
import os
import xml.etree.ElementTree as ET

# Load configuration
CONFIG_FILE = "config.json"
EXTRACTED_URLS_FILE = "extracted_urls.json"

with open(CONFIG_FILE, "r", encoding="utf-8") as f:
    config = json.load(f)

SITEMAP_PATHS = config["standard_sitemap_paths"]
MAIN_DOMAINS = config["main_domains"]

def load_cached_urls():
    """Loads previously extracted URLs from JSON file if it exists."""
    if os.path.exists(EXTRACTED_URLS_FILE):
        with open(EXTRACTED_URLS_FILE, "r", encoding="utf-8") as f:
            return json.load(f)
    return {}

def save_urls_to_cache(urls):
    """Saves extracted URLs and crawling status to JSON."""
    with open(EXTRACTED_URLS_FILE, "w", encoding="utf-8") as f:
        json.dump(urls, f, indent=4)

def extract_urls_from_sitemap(sitemap_url, domain):
    """Extracts URLs from a sitemap.xml file, including multi-level (indexed) sitemaps."""
    try:
        response = requests.get(sitemap_url, timeout=5)
        response_text = response.text.strip()

        # Check if the response is XML (basic validation)
        if not response_text.startswith("<?xml"):
            print(f"⚠️ Warning: {sitemap_url} does not appear to be valid XML. Marking as 'needs_review' and enabling crawling.")
            mark_domain_for_review(domain, sitemap_url, "Invalid XML format", force_crawl=True)
            return []

        try:
            tree = ET.ElementTree(ET.fromstring(response_text))
            root = tree.getroot()

            namespace = {"ns": "http://www.sitemaps.org/schemas/sitemap/0.9"}

            # Check if this is a **Sitemap Index** (contains links to other sitemaps)
            sitemap_entries = root.findall("ns:sitemap/ns:loc", namespace)
            if sitemap_entries:
                print(f"📌 Found sitemap index: {sitemap_url}. Extracting sub-sitemaps...")
                all_urls = []
                for sitemap in sitemap_entries:
                    sub_sitemap_url = sitemap.text.strip()
                    print(f"➡️ Processing sub-sitemap: {sub_sitemap_url}")
                    all_urls.extend(extract_urls_from_sitemap(sub_sitemap_url, domain))  # Recursive call
                return all_urls

            # Otherwise, it's a normal sitemap → Extract URLs
            urls = [elem.text.strip() for elem in root.findall("ns:url/ns:loc", namespace)]
            print(f"✅ Extracted {len(urls)} URLs from sitemap: {sitemap_url}")
            return urls

        except ET.ParseError as e:
            print(f"❌ XML Parsing Error in {sitemap_url}: {e}. Marking as 'needs_review' and enabling crawling.")
            mark_domain_for_review(domain, sitemap_url, str(e), force_crawl=True)
            return []

    except requests.RequestException as e:
        print(f"⚠️ Failed to retrieve sitemap: {sitemap_url}, Error: {e}. Marking as 'needs_review' and enabling crawling.")
        mark_domain_for_review(domain, sitemap_url, str(e), force_crawl=True)
        return []

def mark_domain_for_review(domain, sitemap_url, error_message, force_crawl=False):
    """Marks a domain as 'needs_review' and ensures crawling is enabled in extrahierte_urls.json."""
    if os.path.exists(EXTRACTED_URLS_FILE):
        with open(EXTRACTED_URLS_FILE, "r", encoding="utf-8") as f:
            extracted_urls = json.load(f)
    else:
        extracted_urls = {}

    # Mark the domain for review and enable crawling
    extracted_urls[domain] = {
        "urls": [],
        "crawl": True if force_crawl else False,  # Always set crawl=True when needs_review is true
        "needs_review": True,
        "sitemap_url": sitemap_url,
        "error": error_message
    }

    # Save updated JSON
    with open(EXTRACTED_URLS_FILE, "w", encoding="utf-8") as f:
        json.dump(extracted_urls, f, indent=4)

    print(f"🛑 Domain '{domain}' marked for manual review and set to 'crawl: true' in '{EXTRACTED_URLS_FILE}'.")


# **🔹 Start Sitemap Extraction Process & Store Status**
def main():
    extracted_urls = load_cached_urls()

    for domain in MAIN_DOMAINS:
        print(f"🌍 Processing {domain}...")

        if domain in extracted_urls:
            print(f"✅ Using cached data for {domain}")
        else:
            sitemap_url = get_sitemap(domain)
            if sitemap_url:
                urls_to_scrape = extract_urls_from_sitemap(sitemap_url, domain)
                extracted_urls[domain] = {"urls": urls_to_scrape, "crawl": False}
            else:
                extracted_urls[domain] = {"urls": [], "crawl": True}  # Mark for crawling
            
            save_urls_to_cache(extracted_urls)

    print("✅ Sitemap processing completed. Ready for scraping.")

if __name__ == "__main__":
    main()
```
